```python


# Filter for numeric columns
numeric_cols = data_frame.select_dtypes(include='number').columns

# Dynamically calculate rows and columns for subplots
n_cols = 5
n_rows = -(-len(numeric_cols) // n_cols)  # Ceiling division to get rows

fig, ax = plt.subplots(ncols=n_cols, nrows=n_rows, figsize=(20, 11))
ax = ax.flatten()  # Flatten the axes array for easier indexing

# Plot boxplots
for index, col in enumerate(numeric_cols):
    sns.boxplot(y=col, data=data_frame, ax=ax[index])

# Hide unused subplots
for i in range(len(numeric_cols), len(ax)):
    ax[i].set_visible(False)

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
plt.show()
```
